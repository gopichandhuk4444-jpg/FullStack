```python
from sqlalchemy import create_engine,MetaData,Table,Column,Integer,String,insert
```

```python
engine=create_engine('postgresql+psycopg2://postgres:1533@localhost:5432/pandas_db',echo=True)
```

```python
md_obj=MetaData()
```

```python
user_table=Table(
    'user_',
    md_obj,
    Column('id',Integer,primary_key=True),
    Column('name',String(30),nullable=False)
)
```

```python
md_obj.create_all(engine)
```

```python
print(insert(user_table))
```

```python
with engine.connect() as conn:
    result = conn.execute(
        insert(user_table),
        [
            {"name": "sandy"},
            {"name": "patrick"},
        ],
    )
    conn.commit()
```

```python
md_obj.remove(md_obj.tables['user_'])
```
